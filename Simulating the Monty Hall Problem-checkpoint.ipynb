## Simulating and Visualizing the Monty Hall Problem

Numpy is going to be used as our random number generator, so it needs to be imported

```python
import numpy as np
```

Next, we want want to set up simple counters and the number of trials. At the end, we can use division to get the win percentage for staying and switching. You can write this code without both of the counters provided below, but it is provided more for clarity than anything.

```python
stay_wins = 0
switch_wins = 0
trials = 10000
```

Next we want to create the loop that we will use to run the trials. Within this loop we will generate two random numbers between one and three inclusive. One of these represents the chosen door and the other represents the prize door. If the choosen door is the same as the prize door, then staying would win. If they are different, than switching will always win.

Hopefully this is starting to make sense. You have a 1/3 chance of choosing correctly and a 2/3 chance of choosing incorrectly. If you choose incorrectly first and switch after Monty Hall removews a door, you will ALWAYS win.

```python
for n in range(trials):
    # Numpy's randint accepts the first parameter inclusive and the second parameter exclusively.
    # Below, np.random.randint(1,4) will generate an integer ths is either 1, 2 or 3
    chosen_door = np.random.randint(1,4)
    prize_door = np.random.randint(1,4)
    
    # If staying would have won, we count it towards stay_wins.
    # Otherwise, if the initial guess is incorrect, switching will ALWAYS win. 
    if chosen_door == prize_door:
        stay_wins += 1
    else:
        switch_wins += 1
```

Finally we calculate the win percentage for stay versus switching and display it.

```python
stay_win_pct = stay_wins / trials
switch_win_pct = switch_wins / trials

print('Stay Win Percentage: ', stay_win_pct, '  Switch Win Percentage: ', switch_win_pct)
    
```

And that is all there is to it. Solving this problem analytically using Bayes Rule can, for some, be a bit confusing.Blah blah blah blah

## A Modified Monty Hall Problem

To help understand why switching is always going to give a better probability of winning, we can expand the problem to include more doors. Instead of three doors, let's say we are given a choice of 100 doors initially. After your first choice, Monty Hall removes all but one door. The same rules and assumptions apply. Monty Hall only removes worthless doors and does not touch the prize door or your initial choice. Do you switch? 

We can easily modify the code above to run this kind of simulation. As before, if your initial choice is the same as the prize door, you win by staying. Otherwise, you would have won by switching. The only parts that need to be changed are the parameters inside the randint function.

```python
stay_wins = 0
switch_wins = 0
trials = 100000

for n in range(trials):
    # Numpy's randint accepts the first parameter inclusive and the second parameter exclusively.
    chosen_door = np.random.randint(1,101)
    prize_door = np.random.randint(1,101)

    # If staying would have won, we count it towards stay_wins.
    # Otherwise, if the initial guess is incorrect, switching will ALWAYS win. 
    if chosen_door == prize_door:
        stay_wins += 1
    else:
        switch_wins += 1
        
stay_win_pct = stay_wins / trials
switch_win_pct = switch_wins / trials

print('Stay Win Percentage: ', stay_win_pct, '  Switch Win Percentage: ', switch_win_pct)
```

As more doors are added, the initial probability of choosing correctly the first time decreases. However, because Monty Hall knows which doors are worthless, nearly all of the non-prize doors get removed. It starts to make sense that we should probably switch, as this increases are probability of success greatly!

## Displaying the Results as a Simple Pie Graph

First we need the correct import statement so we can access the matplotlib graphing packages.

```python
import matplotlib.pyplot as plt
```

We will create a "labels" list and "win rates" list to be used as the data for the pie graph. We can also choose some options like color and use the explosion function to seperate the slices a bit more.

```python
labels = ['Stay Win Rate', 'Switch Win Rate']
win_rates = [stay_win_pct, switch_win_pct]

colors = ['steelblue', 'yellow']
# I personally think 0.02 is enough, but you should play around with it to see how it changes the overall graph
explode = (0.02, 0.02)
```

Finally, we put this all into a pie graph. However there is a lot going on here. Starting from the top:

fig, ax = plt.subplots() - This essentially initializes the plot for us. If you can imagine, "fig" is like the window that will contain the plot. "ax" is one of these plots. plt.subplot can be used to have multiple "ax" plots inside of a "fig" window.

ax.axis('equal') - This just guarentees we will have a circle pie graph and not an ugly ovaly one. Comment it out to see how it changes the graph!

ax.set_title("Monty Hall Simulation\n"+"Number of Trials: " + str(trials)) - This part just sets the title. It works very similarly to a print statement, but it can mix strings and numerical values, so trials needs to be convereted.

ax.pie(win_rates, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle = 90) - This essentially draws the pie graph onto our "ax" from above. All of the data, labels, and options get input. The autopct displays the actual percentage from win_rates. Finally, startangle is arbitary. I chose 90 degrees because I liked how it looked.

plt.show() - This displays everything you put together!


```python
fig, ax = plt.subplots()
ax.axis('equal')
ax.set_title("Monty Hall Simulation\n"+"Number of Trials: " + str(trials))
ax.pie(win_rates, explode=explode, colors = colors, labels=labels, autopct='%1.1f%%', shadow=True, startangle = 90)

plt.show()
```

Beautiful! You should definenetly try changing some of the graph options to get a better understanding of how it all works.

## Animating the Simulations

The import statements we need for this are a bit more specialized. First I will go through how to display this animation on any python IDE, and then we will look at displaying them on a Jupyter Notebook such that it works even on a webpage. For most IDEs, the below is sufficient. 

```python
from matplotlib.animation import FuncAnimation
```

Similar to what we did before, the below is just a reinitialization of some of the options we chose originally. This is so you don't have to scroll so far up to change them.

One special note, when fig, ax = plt.subplots() is run, it creates a blank window and blank axes. The %%capture just supresses this. You can comment it out to see what it does. This does not change anything overall, but things look cleaner without it there.

```python
%%capture
trials = 1000
colors = ['steelblue', 'yellow']
explode = (0.02, 0.02)
labels = ['Stay Win Rate', 'Switch Win Rate']
win_rates = [0, 0]

fig, ax = plt.subplots()
```

The next part looks more complicated, but it is not. This function is used to run a single trial and to update our graph so we can see our pie graph animate. The top half of the function is just our original simulation code. The bottom half are just the option we used to create our first pie graph. You'll notice ,however, that there is now an ax.clear() function, which just erases our old graph. The new one is set up afterwards with the same options as before, but with new win_rates data.

```python
def run_trial(trial_num): #trial num starts at 0 and increases by 1 each time update is called   
    #Start of single trial
    chosen_door = np.random.randint(1,4)
    prize_door = np.random.randint(1,4)
    if chosen_door == prize_door:
        win_rates[0] += 1
    else:
        win_rates[1] += 1
    #End of Trial
    ax.clear()
    ax.axis('equal')
    title = str(trial_num+1) #Just used for the title of trial.
    ax.pie(win_rates, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
    ax.set_title("Monty Hall Simulation\n"+"Number of Trials: " + title)
```

Finally, we run the animation function. For the parameters, "fig" is what we are animating, run_trial is the update function, frames is the number of times we will run the trial, and repeat is set to false so that it ends after it finishes all the trials.

```python
pie_animation = FuncAnimation(fig, run_trial, frames = trials, repeat = False)
```

```python
plt.show()
```

Normally, this would work. However Jupyter Notebooks don't like animations. If you run the above code in a python IDE, it should be fine provided you have the correct libraries.

To work around this, we will instead create an HTML video of our animation. We will use a package which will allow us to convert our above animation into something more animated.

```python
from IPython.display import HTML
```

The last step is calling the function we need to create our video. Depending on your number of trials, this can take some time. I suggest starting with 10 trials and then increasing it to your liking. 1000 trials takes at least a minute to create

```python
HTML(pie_animation.to_html5_video())
```
